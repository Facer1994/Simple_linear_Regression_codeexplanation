```python
#importing pandas library
import pandas as pd
```

```python
#Reading the given dataset via read_csv
dataset= pd.read_csv("Salary_Data.csv")
```

```python
#Retrieving the columns from the given dataset
dataset
```

```python
# split the input values  from the dataset and stored in a separate variable
independent=dataset[["YearsExperience"]]
```

```python
# Now viewing the inputs values present in the variable
independent
```

```python
# split the output variables from the dataset and stored in a separate variable
dependent= dataset[["Salary"]]
```

```python
# Now viewing the output values present in the variable
dependent
```

```python
# import train_test_split from sklearn.odel_selection package
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Import linear regression from from sklearn.linear_model package/ library
from sklearn.linear_model import LinearRegression
#Assigning Linearregression function to the variable
regressor=LinearRegression()
#Fit the model with input and output training data
regressor.fit(x_train,y_train)
```

```python
# Now checking the weight present in the created  model
weight= regressor.coef_
weight
```

```python
# Now checking the bias/origin present in the created  model
bias= regressor.intercept_
bias
```

```python
# Assiging the predicted value to a value
y_predict=regressor.predict(x_test)
```

```python
# Importing r2-score from sklearn.metrics package/library
from sklearn.metrics import r2_score
# Calculating r2 score using actual value, predicted value. 
r_score= r2_score(y_test, y_predict) 
```

```python
#Now printing the r2score value by calling the r_score variable
r_score
```

```python
#importing pickle for saving the model
import pickle
#Assigning filename to the varaiable
filename="finalize_model_linear.sav"
```

```python
#Now saving the model into the file using dump function
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Now reading the saved model from the file using load.
loaded= pickle.load(open("finalize_model_linear.sav",'rb'))
```

```python
# Now testing the saved model using predict function by giving some values.
result= loaded.predict([[0]])
```
